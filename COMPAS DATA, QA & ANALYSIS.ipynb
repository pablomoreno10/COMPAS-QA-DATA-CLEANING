## Clean Up Categorical Features For Data Profile

```python
fix_sex_dict={'Male' :0, 'Female' :1}
cp2['gender1']=cp2.Sex_Code_Text.map(fix_sex_dict)
cp2[['Sex_Code_Text','gender1']].groupby(['Sex_Code_Text'])['gender1'].value_counts()
```

```python
cp2['MaritalStatus'].value_counts()
```

```python
def encode_marital_status_binary(cp):
    if cp.MaritalStatus in ["Married", "Significant Other"]:
        return 1
    else:
        return 0

cp2['MaritalStatusBinary'] = cp2.apply(encode_marital_status_binary, axis=1)
cp2.head()
```

```python
cp2['Ethnic_Code_Text'].value_counts()
```

```python
def encode_ethnicity(cp):
    if cp.Ethnic_Code_Text in ["African-American", "African-Am"]:
        return 0
    elif cp.Ethnic_Code_Text == "Caucasian":
        return 1
    elif cp.Ethnic_Code_Text == "Hispanic":
        return 2
    elif cp.Ethnic_Code_Text in ["Asian", "Oriental"]:
        return 3
    elif cp.Ethnic_Code_Text == "Other":
        return 4
    elif cp.Ethnic_Code_Text == "Native American":
        return 5
    elif cp.Ethnic_Code_Text == "Arabic":
        return 6
    else:
        return 99  # Default value for unexpected categories

cp2['EthnicCodeEncoded'] = cp2.apply(encode_ethnicity, axis=1)
cp2.head()
```

```python
cp2['CustodyStatus'].value_counts()
```

```python
custody_status_mapping = {
    'Jail Inmate': 0,
    'Probation': 1,
    'Pretrial Defendant': 2,
    'Residential Program': 3,
    'Parole': 4,
    'Prison Inmate': 5
}

# Apply the mapping
cp2['CustodyStatusEncoded'] = cp2['CustodyStatus'].map(custody_status_mapping)
```

# Answer the following questions prior to your merge
how many columns and rows did you expect in your final file? Why? What type of join/merge did you use (inner or outer, left or right)


1. How many columns and rows did you expect in your final file? Why?

Initially, I used an inner join to merge the compas_raw and two_year_recidivism datasets based on the keys FirstName, LastName, DateOfBirth, and Sex_Code_Text. This resulted in a final dataset that only included individuals present in both datasets, which meant only those who re-offended were retained. While this approach ensured that all rows in the merged dataset had corresponding matches, it excluded non-reoffending individuals, which introduced bias and limited the ability to evaluate the ROR (Release on Recognizance) model's performance across the full population.

To ensure a comprehensive analysis, I switched to an outer left join, which retained all rows from compas_raw (18,427 rows) while adding matching information from two_year_recidivism (4,743 rows). This approach preserved non-reoffending individuals in the final dataset, making it possible to evaluate the ROR model's performance fairly and without bias.

The final merged dataset consisted of 18,427 rows and 90 columns, providing a complete view of both reoffending and non-reoffending individuals for analysis.

2. What type of join/merge did you use (inner, outer, left, or right)?

I initially used an inner join to merge the datasets, but after reviewing the limitations of this approach, I switched to an outer left join. This type of join ensured that all rows from the compas_raw dataset (18,427 rows) were retained in the final merged file, regardless of whether there was a matching row in the two_year_recidivism dataset (4,743 rows). The outer left join preserved non-reoffending individuals, allowing for a comprehensive evaluation of the dataset and the ROR model's performance. The final merged file contained 18,427 rows and 90 columns.
